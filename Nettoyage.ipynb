# Gestion des valeurs manquantes pour les valeurs cibles :

```python
data.dropna(subset=['GHGEmissions(MetricTonsCO2e)'], inplace=True)
```

```python
display(data.describe().transpose())
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(data[data['SiteEnergyUse(kBtu)'] == 0])
```

On voit que pour la grande majorité des mesures où l'énergie consommée par le site est nulle, elle présente également des caractéristiques particulières sur la variables 'ComplianceStatus', synonyme d'erreurs de saisie. On retirera donc ces mesures du dataset.

```python
data = data[data['SiteEnergyUse(kBtu)'] > 0]
```

```python
display(data.describe().transpose())
```

# Imputation aux variables d'usage :

Les variables d'usage peuvent potentiellement avoir un fort impact sur la consommation d'énergie des batiments. Le choix suivant a été fait pour leur imputer des valeurs :

- Pour l'usage principal, s'il n'est pas renseigné, on lui assigne la valeur de 'PrimaryPropertyType' et la surface de batiment.
- Pour les usages secondaires et tertiaires si la valeur est manquante on leur imputera la valeur 'None' et la surface 0.

```python
data['LargestPropertyUseType'].fillna(data['PrimaryPropertyType'], inplace=True)
data['LargestPropertyUseTypeGFA'].fillna(data['PropertyGFABuilding(s)'], inplace=True)

data['SecondLargestPropertyUseType'].fillna('None', inplace=True)
data['SecondLargestPropertyUseTypeGFA'].fillna(0, inplace=True)

data['ThirdLargestPropertyUseType'].fillna('None', inplace=True)
data['ThirdLargestPropertyUseTypeGFA'].fillna(0, inplace=True)

data['ListOfAllPropertyUseTypes'].fillna(data['PrimaryPropertyType'], inplace=True)
```

```python
display(data.describe().transpose())
```

Elimination de redondance dans la variable 'PrimaryPropertyType'

```python
display(data['PrimaryPropertyType'].value_counts())
```

```python
replace_dict = {'sps-district k-12': 'k-12 school', 'distribution center\n':'distribution center',
               'supermarket / grocery store':'supermarket/grocery store',
               'self-storage facility\n':'self-storage facility',
               'college/university':'university', 'residence hall':'residence hall/dormitory',
               'restaurant\n':'restaurant','office':'large office'}
data['PrimaryPropertyType'].replace(to_replace=replace_dict, inplace=True)
```

```python
display(data['PrimaryPropertyType'].value_counts())
```

# Imputation aux variables énergétiques secondaires :

Pour les variables énergétiques 'Weather normalized' il n'y a qu'une seule valeur manquante.
On leur imputera simplement leur équivalents non normalisés.

```python
data['SiteEnergyUseWN(kBtu)'].fillna(data['SiteEnergyUse(kBtu)'], inplace=True)
data['SiteEUIWN(kBtu/sf)'].fillna(data['SourceEUI(kBtu/sf)'], inplace=True)
```

Pour la variable 'ENERGYSTARScore' :

```python
data['ENERGYSTARScore'].fillna(0, inplace=True)
```

# Imputation autres

Pour le nombre d'étage, si la valeur est manquante on considère qu'il n'y en a pas.

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(data.loc[data['NumberofFloors'].isnull()])
```

```python
data['NumberofFloors'].fillna(0, inplace=True)
```

Pour le ZipCode, on regarde le batiment le plus proche à partir de la longitude et la latitude et on lui associe le ZipCode de ce dernier.

```python
def look_closest_zip(df, lat, long):
    lat_distance = df['Latitude'] - lat
    long_distance = df['Longitude'] - long
    tempo = np.sqrt(np.square(lat_distance) + np.square(long_distance))
    closest_zip = df['ZipCode'][tempo.loc[tempo!=0].idxmin()]
    return closest_zip
```

```python
#data['ZipCode'].replace('', np.nan, inplace=True)
data['ZipCode'] = data.apply(
    lambda row: look_closest_zip(data,row['Latitude'],row['Longitude']) if np.isnan(row['ZipCode']) else row['ZipCode'],axis=1)
```

```python
data['PropertyGFAParking'] = abs(data['PropertyGFAParking'])
```
